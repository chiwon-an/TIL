```python
"""
Kafka 컨슈머를 활용하여 메시지를 수신하는 실습입니다.

TODO:
1. kafka-python 라이브러리를 사용하여 KafkaConsumer를 생성합니다.
2. test-topic을 구독하여 메시지를 소비합니다.
3. 수신한 메시지를 출력하며 컨슈머의 동작을 확인합니다.
4. 컨슈머가 지속적으로 실행되도록 설정합니다.
"""

from kafka import KafkaConsumer
```

```python
# TODO 1: KafkaConsumer를 생성하세요.
consumer = KafkaConsumer(
    "test-topic",  # TODO: 구독할 토픽을 설정하세요.
    bootstrap_servers="localhost:9092",  # TODO: Kafka 브로커 주소를 설정하세요.
    auto_offset_reset="earliest",  # TODO: 메시지를 어디서부터 읽을지 설정하세요.
    enable_auto_commit=True,
    group_id="test-group"  # TODO: 컨슈머 그룹 ID를 설정하세요.
)
```

```python
# TODO 2: 메시지를 지속적으로 소비하도록 설정하세요.
for message in consumer:  # 메시지를 소비하는 컨슈머 객체
    print(f"Received: key={message.key}, value={message.value.decode('utf-8')}")
```
