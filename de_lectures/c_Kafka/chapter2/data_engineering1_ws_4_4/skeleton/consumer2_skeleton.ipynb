```python
"""
Kafka 컨슈머 그룹을 활용하여 여러 개의 컨슈머가 메시지를 분산 소비하는 실습입니다.

TODO:
1. kafka-python 라이브러리를 사용하여 여러 개의 KafkaConsumer를 생성합니다.
2. 동일한 컨슈머 그룹 ID를 사용하여 `test-topic`을 구독합니다.
"""

from kafka import KafkaConsumer
```

```python
# TODO 1: KafkaConsumer 인스턴스를 생성하세요.
consumer = KafkaConsumer(
    'test-topic',  # TODO: 구독할 토픽을 설정하세요.
    bootstrap_servers='localhost:9092',  # TODO: Kafka 브로커 주소를 설정하세요.
    auto_offset_reset='earliest',  # TODO: 메시지를 어디서부터 읽을지 설정하세요.
    enable_auto_commit=True,
    group_id='test-topic'  # TODO: 동일한 컨슈머 그룹 ID를 설정하세요.
)
```

```python
# TODO 2: 메시지를 지속적으로 소비하도록 설정하세요.
for message in consumer:  # 메시지를 소비하는 컨슈머 객체
    print(f"Consumer in group 'test-group' received: {message.value.decode('utf-8')}")
```
